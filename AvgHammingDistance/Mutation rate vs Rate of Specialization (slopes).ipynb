```octave
mus = [9 8 7 6 5 4 3 2];
mu = 10.^(-mus);
close all;
specialization = nan(1, length(mus));
ci_dist_specialization = nan(1, length(mus));
tick = 0;
for i = mus
    tick= tick+1;
    C = strcat('Hamming Distance PolyFit mu= 10^(-', num2str(i),') constant Env final.mat');
    load(C);
    eval(sprintf('specialization(1,%d) = average_Slope;', tick));
    eval(sprintf('ci_dist_specialization(1,%d) = alpha*se_avgs;', tick));
end
figure
[-1.*mus', specialization']
y = log(abs(specialization));
[-1.*mus', log(abs(specialization))']
x_lines= mus;
y_lines=y;
%y_lines=specialization;
%plot(x_lines,y_lines);
hold on;
scatter(1:length(mus),y)
%loglog(mu,specialization)
errorbar(1:length(mus),y, ci_dist_specialization,'.')
a =  'Mutation rate vs Rate of Specialization ' ;
title(a, 'FontSize', 10);
xlabel('log(Mutation Rate)', 'FontSize', fontSize);
ylabel('log(Rate of Overspecialization)', 'FontSize', fontSize);
%set(gca,'xtickLabel',{'0','-9','-8','-7','-6','-5', '-4','-3','-2'});
```
