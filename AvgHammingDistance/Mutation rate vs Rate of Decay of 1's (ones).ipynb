```octave
mus = [9 8 7 6 5 3 2];
close all;
decay = nan(1, length(mus));
ci_dist_decay = nan(1, length(mus));
tick = 0;
for i = mus
    tick= tick+1;
    C = strcat('Hamming Distance PolyFit mu= 10^(-', num2str(i),') constant Env final.mat');
    load(C);
    eval(sprintf('decay(1,%d) = average_ones;', tick));
    eval(sprintf('ci_dist_decay(1,%d) = alpha*se_avgones;', tick));
end
figure
[-1.*mus', decay']
y = log(abs(decay));
[-1.*mus', log(abs(decay))']
x_lines= mus;
y_lines=y;
%y_lines=decay;
%plot(x_lines,y_lines);
hold on;
scatter(1:length(mus),y)
errorbar(1:length(mus),y, ci_dist_decay,'.')
a =  'Mutation rate vs Rate of Decay of Functional Genes ' ;
title(a, 'FontSize', 10);
xlabel('log(Mutation Rate)', 'FontSize', fontSize);
ylabel('log(Decay of Functional Genes )', 'FontSize', fontSize);
set(gca,'xtickLabel',{'0','-9','-8','-7','-6','-5','-3','-2'});
```
